```python
#  10.    .
# 1
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#    x   f(x)
x = sp.symbols('x')
#f = sp.cos(3*x - 1) + x
f = sp.exp(sp.cos(2*x))

#    
f1 = sp.diff(f, x)
f2 = sp.diff(f1, x)
f3 = sp.diff(f2, x)
f4 = sp.diff(f3, x)

#  
print("f'(x) =", f1)
print("f''(x) =", f2)
print("f'''(x) =", f3)
print("f''''(x) =", f4)

#         x=0
x0 = 0
f_x0 = f.subs(x, x0).evalf()
f1_x0 = f1.subs(x, x0).evalf()
f2_x0 = f2.subs(x, x0).evalf()
f3_x0 = f3.subs(x, x0).evalf()

#       x=0
T = f_x0 + f1_x0*(x-x0) + (f2_x0/2)*(x-x0)**2 + (f3_x0/6)*(x-x0)**3

#            x=0
print("f(0) =", f_x0.round(3))
print("T(x) =",T.evalf())

#  
x_vals = np.linspace(-0.5, 0.5, 1000)
f_vals = np.array([f.subs(x, xi).evalf() for xi in x_vals])
T_vals = np.array([T.subs(x, xi).evalf() for xi in x_vals])

fig, ax = plt.subplots()
ax.plot(x_vals, f_vals, label='f(x) = cos(3x - 1) + x', color='blue')
ax.plot(x_vals, T_vals, label=' (3rd order)', color='red', linestyle='--')
ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("     ")
plt.grid(True)
plt.show()
```

```python
# 2
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

#    x
x_sym = sp.symbols('x')
#  
f = sp.sin(2*x_sym + 1) + 2*x_sym
#    
f_1 = sp.diff(f, x_sym)
f_2 = sp.diff(f_1, x_sym)
f_3 = sp.diff(f_2, x_sym)
#  
print("  f'(x):", f_1)
print("  f''(x):", f_2)
print("  f'''(x):", f_3)
#    SymPy   Python
f_func = sp.lambdify(x_sym, f, 'numpy')
f_1_func = sp.lambdify(x_sym, f_1, 'numpy')
f_2_func = sp.lambdify(x_sym, f_2, 'numpy')
f_3_func = sp.lambdify(x_sym, f_3, 'numpy')
#  x0
x0 = 0
#        x0
f_x0 = f_func(x0)
#  
print(f"    x0={x0}: {f_x0}")
#   x
x = np.linspace(-2, 2, 400)
#     
f_x = f_func(x)
taylor_x = f_func(0) + f_1_func(0) * x + (f_2_func(0) * x**2) / 2 + (f_3_func(0) * x**3) / 6
#  
plt.figure(figsize=(10, 6))
plt.plot(x, f_x, label='f(x) = sin(2x + 1) + 2x', color='blue')
plt.plot(x, taylor_x, label=' (3rd order)', color='red', linestyle='--')
plt.legend()
plt.title('     ')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#     approximate_taylor_polynomial
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial

#  
def f(x):
    return np.sin(2*x + 1) + 2*x
x = np.linspace(-2.0, 2.0, num=400)
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label="f(x) = sin(2x + 1) + 2x", color='blue')
degree =15
taylor = approximate_taylor_polynomial(f, 0, degree, 1)
print('taylor=', taylor)
plt.plot(x, taylor(x), label=f"degree={degree}", color='red', linestyle='--' )
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',borderaxespad=0.0, shadow=True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("    , ")
plt.tight_layout()
plt.grid()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial
x = np.linspace(-10.0, 10.0, num=400)
plt.plot(x, f(x), label="f(x) = sin(2x + 1) + 2x", color='blue')
for degree in np.arange(1, 15, step=2):
    sin_taylor = approximate_taylor_polynomial(np.sin, 0, degree, 1,
                                               order=degree + 2)
    plt.plot(x, sin_taylor(x), label=f"degree={degree}")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',
           borderaxespad=0.0, shadow=True)
plt.tight_layout()
plt.axis([-10, 10, -10, 10])
plt.show()
```
