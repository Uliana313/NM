#Метод Ейлера

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return np.sin(x / 7) + y

# Початкові дані
a, b = 0.5, 1.5
h = 0.1
y0 = 0.6

# Кількість кроків
n = int((b - a) / h)

# Масиви
x = np.array([a + i*h for i in range(n + 1)])
y = np.zeros(n + 1)
y[0] = y0

# Метод Ейлера
for i in range(n):
    y[i + 1] = y[i] + h * f(x[i], y[i])

# Округлення значень
y_rounded = np.around(y, 4)

# Вивід значень
print("Метод Ейлера :")
for i in range(n + 1):
    print(f"x={x[i]:.1f}, y={y_rounded[i]:.4f}")

# Побудова графіка
plt.plot(x, y, "o", color="blue", label="Точки")
plt.plot(x, y, "-", color="red", label="Лінія")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Метод Ейлера:")
plt.legend()
plt.show()


```

```python
#Додати перевірку odeint для першого завдання
def f_odeint(y, x):
    return np.sin(x / 7) + y

y_odeint = odeint(f_odeint, y0, x)
y_odeint = y_odeint.flatten()

print("\nПеревірка odeint:")
for i in range(len(x)):
    print(f"x={x[i]:.1f}, y={y_odeint[i]:.4f}")
```

```python
#додати перевірку solve_ivp для перешого завдання
def f_solve_ivp(x, y):
    return np.sin(x / 7) + y

sol = solve_ivp(f_solve_ivp, [a, b], [y0], t_eval=x)
y_solve = sol.y[0]

print("\nПеревірка solve_ivp:")
for i in range(len(x)):
    print(f"x={x[i]:.1f}, y={y_solve[i]:.4f}")
```

#Метод Ейлера-Коші

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def f(x, y):
    return math.sin(x/7) + y

a = 0.5
b = 1.5
h = 0.1
y0 = 0.6

n = int((b - a) / h)
x = np.array([a + i*h for i in range(n+1)])
y = np.zeros(n+1)
y[0] = y0

for i in range(n):
    k1 = f(x[i], y[i])
    y_tent = y[i] + h * k1
    k2 = f(x[i+1], y_tent)
    y[i+1] = y[i] + (h/2)*(k1 + k2)

print("Метод Ейлера-Коші:")
for i in range(n+1):
    print(f"x={x[i]:.1f}, y={y[i]:.4f}")

# Побудова графіка
plt.plot(x, y, "o", color="blue", label="Точки")
plt.plot(x, y, "-", color="red", label="Лінія")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Метод Ейлера-Коші ")
plt.legend()
plt.show()

```

#Розв’язок диф. Рівняння за допомогою scipy.integrate import odeint

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Функція, що повертає dy/dx
def model(y, x):
     return np.sin(x / 7) + y

# Початкова умова
y0 = 0.6

# Значення x
x = np.linspace(0.5, 1.5, 6)


# Розв'язання ODE
y = odeint(model, y0, x)

# Виведення результатів
print('x=', x)
print('y=', y.flatten())  # Перетворення в одновимірний масив для виводу

# Побудова графіка результатів
plt.plot(x, y, marker='o')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Розв’язання диф. рівняння')
plt.grid()
plt.show()
```

#Використання scipy.integrate.solve_ivp
Функція solve_ivp є більш сучасним і гнучким способом розв'язання ODE, ніж odeint. Вона також дозволяє використовувати різні методи інтегрування.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Функція, що повертає dy/dx
def model(x, y):
    return np.sin(x / 7) + y

# Початкова умова
y0 = [0.6]  # Потрібно передати як список

# Значення x
x = np.linspace(0.5, 1.5, 6)

# Розв'язання ODE
sol = solve_ivp(model, [0.5, 1.5], y0, t_eval=x)

# Виведення результатів
print('x=', sol.t)
print('y=', sol.y[0])

# Побудова графіка результатів
plt.plot(sol.t, sol.y[0], marker='o')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Розв’язання диф. рівняння')
plt.grid()
plt.show()
```
